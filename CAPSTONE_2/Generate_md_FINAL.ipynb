```python
df = pd.read_csv('billboard_2000s.csv',encoding = "latin" )
df.head()
```

```python
question_intent_artist = []
question_intent_year = []
question_intent_rank = []
question_intent_lyrics = []

for i in range(len(df.Song)):
    random_question = np.random.choice(question_master_intent_artist,replace=True)
    random_question_with_song_name = re.sub("song name",df.Song[i],random_question)    
    question_intent_artist.append(random_question_with_song_name)
    
    random_question = np.random.choice(question_master_intent_year,replace=True)
    random_question_with_song_name = re.sub("song name",df.Song[i],random_question)   
    question_intent_year.append(random_question_with_song_name)

    random_question = np.random.choice(question_master_intent_rank,replace=True)
    random_question_with_song_name = re.sub("song name",df.Song[i],random_question)   
    question_intent_rank.append(random_question_with_song_name)    
    
    random_question = np.random.choice(question_master_intent_lyrics,replace=True)
    random_question_with_song_name = re.sub("song name",df.Song[i],random_question)   
    question_intent_lyrics.append(random_question_with_song_name)
```

```python
file = open("nlu.md","w")

file.write('## intent: artist')
file.write('\n')
for i in range(len(question_intent_artist)):
    file.write(question_intent_artist[i])
    file.write('\n')
file.write('#end_mark_artist#')
file.write('\n')

file.write('\n')    
file.write('## intent: year')
file.write('\n')
for i in range(len(question_intent_year)):
    file.write(question_intent_year[i])
    file.write('\n')
file.write('#end_mark_year#')
file.write('\n')

file.write('\n')    
file.write('## intent: ranking')
file.write('\n')
for i in range(len(question_intent_rank)):
    file.write(question_intent_rank[i])
    file.write('\n')
file.write('#end_mark_ranking#')
file.write('\n')
    
file.write('\n')    
file.write('## intent: lyrics')
file.write('\n')
for i in range(len(question_intent_lyrics)):
    file.write(question_intent_lyrics[i])
    file.write('\n')
file.write('#end_mark_lyrics#')
file.write('\n')

file.close() 
```
